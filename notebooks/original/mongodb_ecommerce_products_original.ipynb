```python
from pymongo import MongoClient

# Conexão com o MongoDB (alterar o URI se necessário)
client = MongoClient("mongodb://localhost:27017/")

# Criando o banco de dados e a coleção
db = client["ecommerce_db"]
products_collection = db["products"]

# Função para inserir produtos
def insert_product(product_id, name, category, price, stock):
    product = {
        "product_id": product_id,
        "name": name,
        "category": category,
        "price": price,
        "stock": stock
    }
    result = products_collection.insert_one(product)
    print(f"Produto inserido com ID: {result.inserted_id}")

# Função para atualizar um produto
def update_product(product_id, updated_data):
    result = products_collection.update_one(
        {"product_id": product_id},
        {"$set": updated_data}
    )
    print(f"Produtos atualizados: {result.modified_count}")

# Função para excluir um produto
def delete_product(product_id):
    result = products_collection.delete_one({"product_id": product_id})
    print(f"Produtos excluídos: {result.deleted_count}")

# Função para buscar produtos
def search_products(category=None):
    query = {}
    if category:
        query["category"] = category
    products = products_collection.find(query)
    for product in products:
        print(product)

# Exemplo de uso
if __name__ == "__main__":
    # Inserindo produtos
    insert_product(1, "Smartphone", "Electronics", 699.99, 50)
    insert_product(2, "Laptop", "Electronics", 999.99, 30)
    insert_product(3, "Headphones", "Accessories", 49.99, 100)

    # Atualizando um produto
    update_product(1, {"price": 649.99, "stock": 45})

    # Buscando produtos da categoria "Electronics"
    print("\nProdutos na categoria 'Electronics':")
    search_products(category="Electronics")

    # Excluindo um produto
    delete_product(3)

    # Buscando todos os produtos
    print("\nTodos os produtos disponíveis:")
    search_products()
```
