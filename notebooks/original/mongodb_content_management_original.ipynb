```python
from pymongo import MongoClient
from datetime import datetime

# Conexão com o MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Criando o banco de dados e a coleção
db = client["content_management_db"]
content_collection = db["contents"]

# Função para inserir conteúdo
def insert_content(content_id, title, content_type, description, created_by, tags=None, metadata=None):
    content = {
        "content_id": content_id,
        "title": title,
        "content_type": content_type,  # e.g., "article", "image", "video"
        "description": description,
        "created_by": created_by,
        "tags": tags if tags else [],
        "metadata": metadata if metadata else {},
        "created_at": datetime.now()
    }
    result = content_collection.insert_one(content)
    print(f"Conteúdo inserido com ID: {result.inserted_id}")

# Função para atualizar o conteúdo
def update_content(content_id, updates):
    result = content_collection.update_one(
        {"content_id": content_id},
        {"$set": updates}
    )
    print(f"Conteúdos atualizados: {result.modified_count}")

# Função para excluir conteúdo
def delete_content(content_id):
    result = content_collection.delete_one({"content_id": content_id})
    print(f"Conteúdos excluídos: {result.deleted_count}")

# Função para buscar conteúdo com filtros
def search_contents(content_type=None, tag=None):
    query = {}
    if content_type:
        query["content_type"] = content_type
    if tag:
        query["tags"] = tag
    contents = content_collection.find(query)
    print("\nResultados da busca:")
    for content in contents:
        print(content)

# Exemplo de uso
if __name__ == "__main__":
    # Inserindo conteúdos
    insert_content(1, "Guia de Python", "article", "Aprenda Python do zero.", "admin", ["python", "programming"], {"word_count": 1500})
    insert_content(2, "Paisagem Marinha", "image", "Foto de uma praia tropical.", "photographer", ["nature", "beach"], {"resolution": "1920x1080"})
    insert_content(3, "Tutorial de Fotografia", "video", "Como capturar paisagens incríveis.", "videographer", ["photography", "tutorial"], {"duration": "15m"})

    # Atualizando um conteúdo
    update_content(1, {"tags": ["python", "programming", "tutorial"]})

    # Buscando conteúdos do tipo "image"
    search_contents(content_type="image")

    # Buscando conteúdos com a tag "tutorial"
    search_contents(tag="tutorial")

    # Excluindo um conteúdo
    delete_content(3)

    # Buscando todos os conteúdos
    print("\nTodos os conteúdos disponíveis:")
    search_contents()
```
