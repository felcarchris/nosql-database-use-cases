```
pip install cassandra-driver

from cassandra.cluster import Cluster

# Conectando ao Cassandra
cluster = Cluster(['127.0.0.1'])
session = cluster.connect()

# Criando keyspace para o gerenciamento de eventos
session.execute("""
    CREATE KEYSPACE IF NOT EXISTS event_management
    WITH replication = {'class': 'SimpleStrategy', 'replication_factor': '1'}
""")

# Selecionando o keyspace
session.set_keyspace('event_management')

# Criando uma tabela para armazenar informações dos eventos
session.execute("""
    CREATE TABLE IF NOT EXISTS events (
        event_id UUID PRIMARY KEY,
        event_name text,
        event_date date,
        location text,
        description text
    )
""")

# Criando uma tabela para armazenar informações dos participantes
session.execute("""
    CREATE TABLE IF NOT EXISTS participants (
        participant_id UUID PRIMARY KEY,
        event_id UUID,
        name text,
        email text
    )
""")
import uuid
from datetime import date

# Inserindo um evento
event_id = uuid.uuid4()
session.execute("""
    INSERT INTO events (event_id, event_name, event_date, location, description)
    VALUES (%s, %s, %s, %s, %s)
""", (event_id, 'Conferência de Tecnologia', date(2024, 3, 15), 'Londres', 'Conferência sobre as últimas tendências em tecnologia.'))

# Inserindo um participante
participant_id = uuid.uuid4()
session.execute("""
    INSERT INTO participants (participant_id, event_id, name, email)
    VALUES (%s, %s, %s, %s)
""", (participant_id, event_id, 'Alice Silva', 'alice@example.com'))

# Consultando eventos
rows = session.execute("SELECT * FROM events")
for row in rows:
    print(row)
```
